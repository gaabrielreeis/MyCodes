```python
# B2
# Liste (1) o código e (2) o nome dos cinco CONs com o maior número
# de atletas registrados. Informe também (3) a quantidade total de atletas
# para cada CON.

query = """
SELECT con, nome_regiao, COUNT(*) as num_atletas
FROM regiao_con NATURAL JOIN pessoa_regiao
GROUP BY id_regiao
ORDER BY COUNT(*) DESC
LIMIT 5;
"""

df = pd.read_sql_query(query, con)
df
```

```python
# B5.
# Liste (1) o nome dos países e (2) o continente com cidades sede de Jogos
# Olímpicos de Inverno.

query = """
SELECT DISTINCT nome_pais, nome_continente
FROM cidade NATURAL JOIN jogos_cidade NATURAL JOIN jogos
WHERE estacao = 'Inverno';
"""

df = pd.read_sql_query(query, con)
df
```

```python
# B6
# Liste (1) o ano, (2) a cidade e (3) a quantidade de esportes diferentes que
# foram disputados em cada edição dos Jogos Olímpicos de Verão. Ordene
# os resultados da edição mais antiga para a mais recente.

query = """
SELECT ano_jogos, nome_cidade, COUNT(DISTINCT id_esporte) as num_esportes
FROM jogos NATURAL JOIN
     jogos_cidade NATURAL JOIN
     cidade NATURAL JOIN
     jogos_competidor NATURAL JOIN
     competidor_evento NATURAL JOIN evento
WHERE estacao = 'Verão'
GROUP BY id_jogos
ORDER BY ano_jogos;
"""

df = pd.read_sql_query(query, con)
df
```

```python
# B7 # REFAZENDO
# Liste (1) o ano e (2) a quantidade de medalhas (i.e., Ouro, Prata ou
# Bronze) que o Brasil conquistou em edições dos Jogos Olímpicos. Ordene
# os resultados da edição mais antiga para a mais recente. Considere a
# quantidade de medalhas por evento e não por atleta ou esporte.

query = """
SELECT ano_jogos , COUNT(DISTINCT id_evento)
FROM jogos NATURAL JOIN
     jogos_competidor NATURAL JOIN
     competidor_evento NATURAL JOIN
     pessoa NATURAL JOIN
     pessoa_regiao NATURAL JOIN
     regiao_con
WHERE con = 'BRA' AND id_medalha != 4
GROUP BY id_jogos
ORDER BY ano_jogos;
"""

df = pd.read_sql_query(query, con)
df
```

```python
# B8
# Liste (1) o ano dos Jogos Olímpicos de Inverno e (2) a quantidade de
# CONs participantes. Ordene os resultados da edição mais antiga para a
# mais recente.

query = """
SELECT ano_jogos , COUNT(DISTINCT con) as num_CONs
FROM jogos NATURAL JOIN
     jogos_competidor NATURAL JOIN
     pessoa NATURAL JOIN
     pessoa_regiao NATURAL JOIN
     regiao_con
WHERE estacao = 'Inverno'
GROUP BY ano_jogos
ORDER BY ano_jogos;
"""

df = pd.read_sql_query(query, con)
df
```
